```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

We are going to request information from "https://en.wikipedia.org/wiki/List_of_largest_banks"

```python
html_data=requests.get('https://en.wikipedia.org/wiki/List_of_largest_banks').text
soup=BeautifulSoup(html_data, 'lxml')
```

Then, we are going to create a DataFrame with the table of banks

```python
data=pd.DataFrame(columns=["Rank","Name","Market Cap (US$ Billion)"])

for row in soup.find_all('tbody')[3].find_all('tr'):
    col=row.find_all('td')
    if(col != []):
        index=col[0]
        title=col[1].text
        value=col[2]
        data=data.append({"Rank":index, "Name":title, "Market Cap (US$ Billion)": value}, ignore_index=True)
```
