```python
import requests
import pandas as pd
```

It is needed for this project to have a API Key in https://coinlayer.com/signup/free

```python
url = "http://api.coinlayer.com/live?access_key=60d7b6578982c2b4a08d0ae3b5c98a1f"

data=requests.get(url).text

data_json=pd.read_json(data, lines=True)

del(data_json['success'])
del(data_json['terms'])
del(data_json['timestamp'])
del(data_json['privacy'])
del(data_json['target'])

Rates=data_json['rates'][0]

pandas=pd.DataFrame(list(Rates.items()), columns=['Currency', 'Rate'])
```

Storing the information in a csv file

```python
pandas.to_csv('exchange_rates_1.csv')
```
